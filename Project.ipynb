```python
from scipy.stats import ttest_ind

# Supprimer les lignes avec des valeurs manquantes dans les colonnes pertinentes
theses_df = theses_df.dropna(subset=['fr_title', 'buzzwords'])

# Séparer les données en deux groupes en fonction de la présence ou de l'absence de buzzword
group_0 = theses_df[theses_df['buzzwords'] == 0]['fr_title']
group_1 = theses_df[theses_df['buzzwords'] == 1]['fr_title']

# Effectuer le test de Student
t_statistic, p_value = ttest_ind(group_0, group_1)

# Afficher les résultats du test
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Interpréter les résultats
alpha = 0.05
if p_value < alpha:
    print("Il y a une différence significative entre les groupes.")
else:
    print("Il n'y a pas de différence significative entre les groupes.")
plt.figure(figsize=(10, 6))
sns.barplot(x='buzzword_binary', y='fr_title', data=theses_df, ci='sd')
plt.xlabel('Présence de Buzzword (0: Non, 1: Oui)', color="white")
plt.ylabel('Longueur du Titre', color="white")
plt.title('Longueur du Titre en fonction de la Présence de Buzzword', color="white")
plt.gca().tick_params(axis='x', colors='white')
plt.gca().tick_params(axis='y', colors='white')
plt.show()
```

```python
from scipy.stats import chi2_contingency

# Supprimer les lignes avec des valeurs manquantes dans les colonnes pertinentes
theses_df = theses_df.dropna(subset=['buzzwords', 'Institution'])

# Créer un tableau de contingence
contingency_table = pd.crosstab(theses_df['buzzwords'], theses_df['Institution'])

# Effectuer le test du chi carré
chi2, p, _, _ = chi2_contingency(contingency_table)

# Afficher les résultats
print(f"Test du Chi carré : {chi2}")
print(f"P-value : {p}")

# Interpréter les résultats
alpha = 0.05
if p < alpha:
    print("Il existe une association significative entre la présence de buzzwords et l'université.")
else:
    print("Il n'y a pas d'association significative entre la présence de buzzwords et l'université.")
```

```python
import pandas as pd
import folium
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output

# Read the CSV file
filename = '/Users/clara_bourdon/Downloads/df_thesis_updated-2.csv'
df_modif = pd.read_csv(filename)
```

```python
df_modif.head()
```

```python
df_modif= df_modif.dropna(subset=['version_i'])

# Afficher le résultat
df_modif.head()
```

```python

# Calculate the total number of theses per university
total_theses_per_university = df_modif['Institution'].groupby('version_i').count().reset_index(name='total_theses')

# Create city counts DataFrame
city_counts = df_modif['Institution'].groupby(df_modif['version_i']).count().reset_index(name='school_count')

# Create a Folium map of France
france_map = folium.Map(location=[46.6031, 1.8883], zoom_start=6)

# Add markers for each city with a bubble plot
for index, row in city_counts.iterrows():
    try:
        location = tuple(map(float, row['version_i'].split(', ')))

        # Calculate the size of the bubble plot based on the percentage of theses in the university
        size_percentage = (row['school_count'] / total_theses_per_university[total_theses_per_university['version_i'] == row['version_i']]['total_theses'].values[0]) * 100

        folium.CircleMarker(
            location=location,
            radius=size_percentage * 0.5,  # Adjust the scaling factor as needed
            color='blue',
            fill=True,
            fill_color='blue',
            fill_opacity=0.7,
            popup=f"{row['version_i']}<br>{row['school_count']} écoles<br>{size_percentage:.2f}% des thèses"
        ).add_to(france_map)
    except (ValueError, AttributeError, ZeroDivisionError):
        # Handle errors or division by zero
        location = (0, 0)

# Save the map as an HTML file
france_map.save("bubble_plot_france.html")

# Create a Dash web application
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Bubble Plot of Schools in France"),
    
    # Dropdown slicer for buzzwords
    dcc.Dropdown(
        id='buzzwords-dropdown',
        options=[{'label': buzzwords, 'value': buzzwords} for buzzwords in df_modif['buzzwords'].unique()],
        multi=True,
        value=[],  # Initial selected values
        placeholder="Select Buzzwords"
    ),

    # HTML iframe to hold the Folium map
    html.Iframe(id='france-map', width='100%', height='600px'),

    # ...
])

# Define callback to update the map based on selected buzzwords
@app.callback(
    Output('france-map', 'srcDoc'),
    [Input('buzzwords-dropdown', 'value')]
)
def update_map(selected_buzzwords):
    # Filter the DataFrame based on selected buzzwords
    filtered_df = df_modif[df_modif['buzzwords'].isin(selected_buzzwords)]

    # Create city counts DataFrame for the filtered data
    city_counts_filtered = filtered_df['Institution'].groupby(filtered_df['version_i']).count().reset_index(name='school_count')

    # Create a Folium map for the filtered data
    france_map_filtered = folium.Map(location=[46.6031, 1.8883], zoom_start=6)

    # Add markers for each city with a bubble plot for the filtered data
    for index, row in city_counts_filtered.iterrows():
        try:
            location = tuple(map(float, row['version_i'].split(', ')))

            # Calculate the size of the bubble plot based on the percentage of theses in the university
            size_percentage = (row['school_count'] / total_theses_per_university[total_theses_per_university['version_i'] == row['version_i']]['total_theses'].values[0]) * 100

            folium.CircleMarker(
                location=location,
                radius=size_percentage * 0.5,  # Adjust the scaling factor as needed
                color='blue',
                fill=True,
                fill_color='blue',
                fill_opacity=0.7,
                popup=f"{row['version_i']}<br>{row['school_count']} écoles<br>{size_percentage:.2f}% des thèses"
            ).add_to(france_map_filtered)
        except (ValueError, AttributeError, ZeroDivisionError):
            location = (0, 0)

    # Save the filtered map as an HTML file
    filtered_map_filename = "bubble_plot_france_filtered.html"
    france_map_filtered.save(filtered_map_filename)

    return open(filtered_map_filename, 'r').read()

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)






```
